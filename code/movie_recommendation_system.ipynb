```python
def find_mode(arr):
    counts = {}
    for num in arr:
        if num in counts:
            counts[num] += 1
        else:
            counts[num] = 1

    modes = []
    max_count = max(counts.values())

    for num, count in counts.items():
        if count == max_count:
            modes.append(num)
    return modes

find_mode(data['Popularity'])
```

```python
# 1. ML Algorithm
# This code is for movie recommendation using K-Nearest Neighbors (KNN) algorithm.
# KNN is machine learning algorithm that finds closest neighbors to a data point using distance metric.
# Here, it uses cosine similarity between movie overviews to recommend similar movies.
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
```

```python
# Step 1: Prepare the data
data = data.dropna(subset=['Title', 'Overview'])
data['Overview'] = data['Overview'].astype(str)
```

```python
# Step 2: Convert text to numbers using TF-IDF
# TF-IDF (Term Frequency-Inverse Document Frequency) is used to show how important words are in movie overviews.
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(data['Overview'])
```

```python
# Step 3: Train the KNN model
# Use Nearest Neighbors model with cosine similarity as distance metric.
# 'brute' algorithm is used to calculate distances on TF-IDF matrix.
nn_model = NearestNeighbors(metric='cosine', algorithm='brute')
nn_model.fit(tfidf_matrix)
```

```python
# Step 4: Make index for movie titles
# This helps to find movie index fast when making recommendations.
indices = pd.Series(data.index, index=data['Title'].str.lower()).drop_duplicates()
```

```python
# this function takes a movie title and returns a list of recommended movies based on their overviews using knn algorithm
def recommend_movies(title, num_recommendations=5):
    title = title.lower()
    if title not in indices:
        return f"Movie '{title}' not found in dataset."

    idx = indices[title]
    distances, indices_rec = nn_model.kneighbors(tfidf_matrix[idx], n_neighbors=num_recommendations + 1)
    recommended_indices = indices_rec[0][1:]  # Exclude the input movie itself

    return data[['Title', 'Overview']].iloc[recommended_indices]
```

```python
# 2. ML Algorithm
# This algorithm uses clustering to group movies based on their genres and 
# then recommends movies from the same cluster.
from sklearn.cluster import KMeans
```

```python
# Step 1: Preprocess the Genre column
# The 'Genre' column contains a comma-separated list of genres for each movie.
data['Genre_list'] = (
    data['Genre']
      .fillna('') # Replace NaN values with an empty string
      .str.split(',') # Split the genres by commas into a list
      .apply(lambda lst: [g.strip()
                         for g in lst
                         if g.strip()]) # Remove whitespace and empty genres
)

# Step 2: Convert the Genre list into a binary matrix
# We use MultiLabelBinarizer to create a binary matrix where each column represents a genre,
# and each row indicates whether the movie belongs to that genre (1 for yes, 0 for no).
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
genre_matrix = mlb.fit_transform(data['Genre_list']) # Fit and transform the genre list into a binary matrix


# Step 3: Apply KMeans clustering
# We use KMeans to group movies into clusters based on their genres.
# The number of clusters (n_clusters) is set to 5, but this can be adjusted based on the dataset.
# from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, random_state=42) 
data['Genre_Cluster'] = kmeans.fit_predict(genre_matrix) # Fit the model and assign each movie to a cluster

# Step 4: Display the movies in each cluster
# For each cluster, we display the first 5 movies along with their titles and genres.
for cluster_id in sorted(data['Genre_Cluster'].unique()):
    print(f"\nCluster {cluster_id}:")
    display(data.loc[data['Genre_Cluster']==cluster_id, ['Title','Genre_list']].head(5))

```

```python
# The Silhouette Score measures how well each data point is clustered.
# A higher score indicates better-defined clusters.
# Score ranges from -1 to 1, where a score close to 1 indicates that the points are well clustered,
# a score close to 0 indicates overlapping clusters, and a negative score indicates that the points 
# may have been assigned to the wrong cluster.
from sklearn.metrics import silhouette_score

score = silhouette_score(genre_matrix, data['Genre_Cluster'])
print(f"Silhouette Score: {score:.3f}")

# It is normal to have overlapping clusters for movie genres because movies often belong to multiple genres, 
# and genres themselves can have significant overlap in themes or characteristics.
# For example, a movie classified as both "Action" and "Adventure" might share similarities with 
# movies in either genre, leading to overlapping clusters.
# that is why the silhouette score is not very high, indicating that the clusters are not perfectly separated.
```

```python
# function to recommend movies based on genre cluster applying the KMeans clustering algorithm
def recommend_by_genre_cluster(title, n=5):
    title_lower = title.lower()
    matches = data.index[data['Title'].str.lower() == title_lower]
    if len(matches) == 0:
        return f"Movie '{title}' not found."

    idx = matches[0]
    cluster_id = data.at[idx, 'Genre_Cluster']

    peers = data[data['Genre_Cluster'] == cluster_id].drop(idx)
    if peers.empty:
        return f"No other movies in cluster {cluster_id}."

    sample = peers.sample(min(n, len(peers)), random_state=42)
    return sample[['Title', 'Genre_list']]

```

```python
recommend_by_genre_cluster("The Batman", n=5)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
# 3. ML Algorithm
# This algorithm uses a Random Forest Classifier to predict whether a movie is highly rated based on its features.

# Step 1: Preprocess the data
# First we preprocess the data by creating a new binary column 'Highly_Rated'.
# If the 'Vote_Average' is greater than or equal to 7.5, the movie is considered highly rated (value = 1).
# Otherwise it is not highly rated (value = 0).
data['Vote_Average'] = pd.to_numeric(data['Vote_Average'], errors='coerce')
data['Highly_Rated'] = (data['Vote_Average'] >= 7.5).astype(int)

# Step 2: Create a feature set
# Next, we combine numerical features ('Popularity' and 'Vote_Count') with genre features into a single feature set.
# Missing values in numerical features are replaced with 0 to avoid errors during training.
X_num = data[['Popularity', 'Vote_Count']].fillna(0)
X = np.hstack([X_num.values, genre_matrix])
y = data['Highly_Rated'].values # Target variable for classification

# Step 3: Split the dataset into training and testing sets
# We split the dataset into training and testing sets.
# 75% of the data is used for training, and 25% is used for testing.
# Stratify ensures that the proportion of highly rated and not highly rated movies is maintained in both sets.
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

# Step 4: Train the Random Forest Classifier
# A Random Forest Classifier is initialized with 100 decision trees.
# The model is trained using the training data (X_train and y_train).
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Step 5: Evaluate the model
# After training, we use the model to predict the labels for the test set (X_test).
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))

# A confusion matrix is created to visualize the number of correct and incorrect predictions.
cm = confusion_matrix(y_test, y_pred)

# Step 6: Display the confusion matrix
# The confusion matrix is displayed using a heatmap for better visualization.
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=['Not Highly Rated', 'Highly Rated']
)
disp.plot(cmap=plt.cm.Blues, values_format='d')
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.show()


```
